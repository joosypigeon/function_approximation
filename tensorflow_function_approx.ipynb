# Tensorflow implementation of function approx

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Define the function to approximate
def my_function(x):
    return np.sin(5*x)+2*np.cos(10*x**2)

# Generate training inputs and outputs
#x_train = np.random.uniform(0.0, 1.0, 100).reshape(-1, 1)  # inputs
x_train = np.linspace(0.0, 1.0, 100)#.reshape(-1, 1)
y_train = my_function(x_train)  # outputs

# Create a sequential model
model = keras.models.Sequential([
    keras.layers.Dense(32, activation='relu', input_shape=(1,)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1)
])

sgd=tf.keras.optimizers.SGD(learning_rate=0.01)

tb_callback = tf.keras.callbacks.TensorBoard(log_dir="./logs", histogram_freq=1)

model.compile(optimizer=sgd, loss='mse')
epochs = 1000
model.fit(x_train, y_train, batch_size=1, epochs=epochs, verbose=2, callbacks=[tb_callback])
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0.0, 1.0, 100)#.reshape(-1, 1)
y_function = my_function(x)
y_predict = model.predict(x)

# Create the plot
plt.figure(figsize=(8, 6))  # Optional: Specifies the size of the figure
plt.plot(x, y_function, color='red')
plt.plot(x, y_predict, color='blue')

plt.title('Plot of y predicted from x')  # Optional: Gives a title to your plot
plt.xlabel('x')  # Optional: Labels the x-axis
plt.ylabel('y')  # Optional: Labels the y-axis
plt.grid(True)  # Optional: Adds a grid to your plot
plt.show()
```
